## Gráficos de dispersão categóricos

São gráficos de dispersão que permitem avaliar mais de uma categoria ao mesmo tempo em uma mesma visualização(gráfico)

Diferente dos histogramas e do gráfico de densidade, é possível avaliar variáveis categóricas e não apenas quantitativas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

**Analisando `age`, `class`, `sex` e `survived`.**

**Gráfico stripplot**

Esse gráfico permite a visualização dos dados de uma variavel dividindo-as em classe.

Veja que os dados da coluna `class` são do tipo categoricos. O seaborn soma a quantidade de elementos de cada dado diferente, como se criasse um outro dataset agora com o índice derivado dos valores da coluna. 

Em seguida, ele distribui os pontos em colunas por classe no eixo x ou y (dependendo de qual você utiliza para inserir os dados categóricos), enquanto os valores quantitativos, referente a outra variável que está sendo comparada, fica no eixo oposto

Uma característica do `stripplot` é que ele sobrepõe os dados que estão muito próximos, o que pode prejudicar a visão da distribuição quando há muitas amostras 

```python
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe')
sns.stripplot(x='class', y='age', data=titanic)
plt.show()
```

**Análise**

Podemos analisar que a menos pessoas jovens na primeira classe

**Plotando gráfico de enxame(swarmplot**

O gráfico de enxame é uma variação do stripplot, permitindo uma melhor visualização da distribuição dos dados por classe já que não sobrepõe as amostras, diferente do que ocorre no stripplot

```python
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe')
sns.swarmplot(x='class', y='age', data=titanic)
plt.show()
```

**Análise**

Sem a sobreposição podemos observar melhor a distribuição dos pontos entre os valores de idade. 

Na primeira classe há distribuição de pessoas é mais unifor começando na idade de 20 anos para cima. Há poucas pessoas com menos de 20.

Na segunda classe há mais pessoas entre 20 e 35 anos.

Na terceira classe há mais pessoas jovens entre 0 e 30 anos do que nas outras classes

### Analisando distribuição com dois atributos categóricos

**Assim como em outros gráficos, é possível utilizar de parâmetros como `hue` para mapear outra variável categórica, dessa vez através da cor**

---
**Vamos analisar a distribuição de idade em relação a classe ao sexo**

```python
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe e sexo')
sns.stripplot(x=titanic['class'], y=titanic['age'], hue=titanic['sex'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe e sexo')
sns.swarmplot(x=titanic['class'], y=titanic['age'], hue=titanic['sex'])
plt.show()
```

**Análise**

Em ambos os gráficos é possível observar algumas diferenças por classe em relação ao sexo. Mas isso é bem mais visível quando há uma discrepância significativa entre as quantidades.

Em relação ao gráfico `stripplot` é possível perceber que a análise da área quando há uma discrepância grande de dados entre os sexos é mais fácil de ser observado. Isso porque, como exemplo da terceira classe, os pontos em maior quantidade se junto formando um único bloco é absolvido rapidamente pela nossa percepção, o que permite um julgamento rápido de que esses pontos são os "vencedore" naquela classe. **O `stripplot` é melhor para ver densidade quando há grande diferença entre quantidade em variável categóricas representadas por cor**.

Por outro lado, o gráfico de enxame `swarmplot` é útil para mostrar melhor a distribuição, o que permite ver e perceber a densidade dos pontos mais lentamente, mas com maior precisão. Mais devagar porque leva um pouco mais de tempo para percepção absolver a área maior dos pontos, uma vez que esse gráfico não soprepõe dados, e com mais precisão porque é possível observar cada ponto. Com isso, **o gráfico de enxame é útil para ver melhor a distribuição e para diferenciar melhor quando os pontos de cada classe tem quantidade muito próximas**.

### Visualização de gráfico de dispersão categórico em grades

**Vamos avaliar não só a classe e o sexo, mais também quem sobreviveu**

**Usamos o método `catplot` para poder gerar grades de gráficos de dispersão categórico.**

**Esse método é derivado da classe `FaceGrid` que permite fazer visualizações em grades. Ela também possui métodos específicos para caracterizar esse tipo de visualização, não sendo possível utilizar os métodos padrões usados em visualizações anteriores**

**No `catplot` inserimos os atributos que normalmente inserimos no `stripplot` e no `swarmplot` a diferença é que adicionamos o atributo `col` que recebe a variável cuja quantidade de valores dividira em mesma quantidade os gráficos na visualização, e adicionamos o atributo `kind` para definir o tipo de gráfico que queremos, ex: `kind="strip"` (stripplot); `kind="swarm"`(swarmplot)**

```python
titanic.head()
```

```python
a = sns.catplot(x='pclass', y='age', hue='sex', col='survived', kind='strip', data=titanic)
a.fig.suptitle('Sobreviventes por classe', x=0.5, y = 1.2, size= 20, fontweight='black')
plt.show()
```

```python
a = sns.catplot(x='pclass', y='age', hue='sex', col='survived', kind='swarm', data=titanic)
a.fig.suptitle('Sobreviventes por classe', x=0.5, y = 1.2, size= 20, fontweight='black')
plt.show()
```

**Análise**

Observe que há mais sobreviventes mulheres do que homens e a maioria das crianças sobreviveram também. Isso é uma questão histórica daquele evento, já que a regra "Primeiro mulheres e criança" foi considera no momento da retirada do navio 



**Conclusão sobre uso de gráfico `swarm` e `strip`**

O gráfico swarm tem a vantagem de representar melhor a distribuição quando há uma grande quantidade de dados. Entretanto, ele perde precisão em relação a outros gráficos, apesar de ser melhor em relação ao strip, quando se trata de análise de dados mapeados por cor

### Gráfico de caixa

O gráfico de caixa permite uma visualização compacta da distribuição de dados, além de poder representar múltiplas distribuições simultaneamente.

**Conceitos de um gráfico de caixa**

Por se tratar de um gráfico com múltiplas camadas de interpretação, precisamos saber os conceitos para que sua análise se torne menos complexa

- Conceitos
    - Quartis - São valores que dividem um conjunto de dados que está em rol em 4 partes. 
        - 1º - O primeiro quartil é um valor que divide o conjunto de dados: 25% dos valores abaixo dele, 75% acima
        - 2º - O segundo quartil(medianao) é um valor que divide o conjunto de dados: 50% dos valores abaixo dele, 50% acima
        - 3º - O terceiro quartil é um valor que divide o conjunto de dados: 75% dos valores abaixo dele, 25% acima
        
**Os quartis podem ser observados usando o método `describe()` no conjunto de dados em análise**

```python
titanic['age'].describe()
```

**Vamos analisar a parte visual do `boxplot` e compará-lo com um `stripplot`.**



<img src='apresentacao_boxplot.jpg'>

Observe que o `boxplot` é muito mais compacto que o `stripplot`, além disso, sabendo das informações sobre os quartis, podemos extrair e inferir vários dados.

A caixa no centro representa 50% dos dados dos conjunto de dados. A primeira linha é o valor do primeiro quartil(20.125000). A segunda linha representa a mediana (28.0000) e a terceira linha marca o valor do terceiro quartil(38.0000). Sendo assim, se abaixo do primeiro quartil temos os primeiros 25% dos dados e acima do terceiro quartil temos os últimos 25% dos dados, podemos inferir que **dentro da caixa há metade dos dados do conjunto de dados**.

Fora dela, nas linhas, denominadas de **bigodes**, temos, normalmente o restante dos dados, ou seja, os outros 50% dos dados do conjunto. Há exceção quando existe **outliers**, que são dados extremos dentro do conjunto de dados, que se distanciam muito da média dos valores. Esses dados são representados pelos pontos pretos que ultrapassam o limite dos bigodes.

**Mas como é medido esses bigodes e porque o que está fora são outliers ?**

Os bigodes são delimitados tanto pelos valores do primeiro e terceiro quartil, esse intervalo dentro da caixa é conhecido como **amplitude interquartil (AIQ)** quanto pelas delimitações mais externas do bigode, que são dados por (1,5xAIQ), contendo cerca de 99% dos dados. Além dessas limitações, os dados são considerados outliers(dados discrepantes) 

---
**Vamos criar os nossos próprios gráficos de caixa**

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros em cada classe')
sns.boxplot(x='class', y='age', data=titanic)
plt.show()
```

**Análise**

Em comparação com os gráficos `strip` e `swarm` notamos que o destaque ficou pela forma com que os valores das classes são representadas. Não mais como pontos, caixas compactas representam a distribuição dos dados.

Esse tipo de visualização permite ter uma visão mais clara das medidas estatísticas pelos gráficos, além de permitir ver a distribuição dos dados de uma maneira mais comparativa.

Observe que não é possível comparar entre as representações de caixa a quantidade dos valores, onde tem mais valores e onde tem menos valores em cada categoria da variável representada. Isso porque o objetivo do gráfico não é saber onde há mais ou menos e sim comparar as medidas estatísticas

---
**Representando o sexo também**

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros em cada classe por sexo')
sns.boxplot(x='class', y='age', hue='sex', data=titanic)
plt.show()
```

**Análise**

Essa visualização permite ter uma noção comparativa sobre as distribuições de idade por sexo em cada classe, enquanto que a visualização gerada pelo swarmplot() permite apenas comparar a proporção entre os sexos. Por essa visualização, pode-se perceber, por exemplo, que as mulheres são mais jovens do que os homens em todas as classes.

---

**Existe uma variação do boxplot denominado de `boxenplot`. Essa variante permite a visualização da espessura do bigode da caixa, dando uma melhor percepção visual da quantidade de valores nas caudas**

```python
plt.figure(figsize=(15,8))
plt.title('Idade de passageiros em cada classe')
sns.boxenplot(x='class', y='age', hue='sex', data=titanic)
plt.show()
```

**Podemos aplicar o `catplot` com a caracteristica de um `boxplot`: tanto o comum quanto o `boxenplot`.**

```python
a = sns.catplot(x="class", y="age", hue = 'sex', col = 'survived', kind = 'boxen', data=titanic)
a.fig.suptitle('Sobreviventes por classe', x=0.5, y=1.2, size=15, weight='black')
plt.show()
```

**Diferenças entre `boxplot`s e os `stripplot`s**

Perceba que o gráfico de caixa fornece uma melhor noção geral sobre os intervalos das distribuições, onde se encontram os dados e os dados discrepantes. Já o gráfico de enxame fornece uma noção sobre as quantidades e o formato da distribuição, que não é apresentada no gráfico de caixa.

### Gráfico de violino

**É outra variação dos gráficos de dispersão categórico. Dessa vez ele cria uma curva de probabilidade de ocorrência  graça a aplicação da estimativa de densidade de ocorrência (KDE), e a espelha, dando a percepção do gráfico aproximada de um violino**

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros em cada classe')
sns.violinplot(x='class', y='age', data=titanic)
plt.show()
```

Note que o formato dos violinos lembra bastante os gráficos de enxame. Ou seja, com eles, pode-se visualizar as nuances da distribuição que, com os gráficos de caixa, não era possível. Porém, diferente do gráfico de dispersão, o gráfico de violino mostra probabilidades, e não os dados diretamente, como pontos. Isso significa que ele não sofre a limitação de precisar de mais espaço conforme aumentam os dados.

Além disso, o gráfico de violino também mostra uma versão reduzida do gráfico de caixa, representado dentro de cada violino. Dessa maneira, o ponto branco, no meio, representa a mediana; a linha mais espessa representa a caixa; e as linhas menos espessas representam os bigodes.

**Podemos combinar dois gráficos de dispersão diferentes**

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros em cada classe')
sns.violinplot(x="class", y="age", data=titanic)
sns.swarmplot(x="class", y="age",color = 'black', size = 3, data=titanic)
plt.show()
```

**Podemos visualizar mais de 1 variável também**

Primeiro analisamos o sexo em violinplots de separados

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros em cada classe por sexo')
sns.violinplot(x="class", y="age", hue = 'sex', data=titanic)
plt.show()
```

Podemos modificar o atributo `split` para `True`, assim é possível repartir o violino, que deixa de espelhar o outro lado, e passa a representar a probabilidade de ocorrência das variáveis nos lados opostos

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros em cada classe por sexo')
sns.violinplot(x="class", y="age", hue = 'sex', split=True, data=titanic)
plt.show()
```

**Aplicando o `catplot` no gráfico de violino**

```python
a = sns.catplot(x="class", y="age", hue = 'sex', col = 'survived', kind = 'violin', split = True, data=titanic)
a.fig.suptitle('Sobreviventes por classe', x = 0.5, y = 1.1, size=15, weight='black')
plt.show()
```

Esse gráfico chama a atenção para a curva de probabilidade feminina da primeira classe dos não-sobreviventes, indicando que houveram crianças do sexo feminino da primeira classe que não sobreviveram. Perceba que o gráfico de violino, assim como o gráfico de caixa, não indica quantidades. Quando existem muitos dados, isso é desejado, mas quando há um menor número, pode ser mais interessante visualizar com o gráfico de enxame.

Note também que pela curva de probabilidade ser mais achatada e ter um decaimento ou subida mais longo, permite analisar que a probabilidade de ocorrência de todos aqueles valores são bem próximos, ou seja, pode haver uma quantidade similar de valores ao longo da curva bem distribuidos 

```python
a = sns.catplot(x="class", y="age", hue = 'sex', col = 'survived', kind = 'swarm', dodge = True, data=titanic)
a.fig.suptitle('Sobreviventes por classe', x = 0.5, y = 1.1, size=15, weight='black')
plt.show()
```
